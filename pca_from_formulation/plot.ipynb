```python
import numpy as np
import os
import matplotlib.pyplot as plt

relative_diffs_files = [
    "mean_relative_diffs_train_wtrainable.npy",
    "mean_relative_diffs_test_wtrainable.npy",
    "mean_relative_diffs_train_linear.npy",
    "mean_relative_diffs_test_linear.npy",
    "mean_relative_diffs_train_DicWithInputs.npy",
    "mean_relative_diffs_test_DicWithInputs.npy",
    "mean_relative_diffs_train_MatrixWithInputs.npy",
    "mean_relative_diffs_test_MatrixWithInputs.npy",
]

relative_error_files = [
    "mean_relative_error_train_wtrainable.npy",
    "mean_relative_error_test_wtrainable.npy",
    "mean_relative_error_train_linear.npy",
    "mean_relative_error_test_linear.npy",
    "mean_relative_error_train_DicWithInputs.npy",
    "mean_relative_error_test_DicWithInputs.npy",
    "mean_relative_error_train_MatrixWithInputs.npy",
    "mean_relative_error_test_MatrixWithInputs.npy",
]

# Load all relative diffs
relative_diffs = {file: np.load(os.path.join(data_dir, file)) for file in relative_diffs_files}

# Load all relative errors
relative_errors = {file: np.load(os.path.join(data_dir, file)) for file in relative_error_files}
```

```python
name_list = [
    "$A(\nu) \Psi + B(\nu) I_p$ - train$",
    "$A(\nu) \Psi + B(\nu) I_p$ - test",
    "$A(\nu) + B(\nu) G(I_p)$ - train",
    "$A(\nu) + B(\nu) G(I_p)$ - test",
    "$A(\nu) \Psi(I_b, I_p)$ - train",
    "$A(\nu) \Psi(I_b, I_p)$ - test",
    "$A(\nu, I_p) \Psi(I_b)$ - train",
    "$A(\nu, I_p) \Psi(I_b)$ - test",
]

# Plotting all relative diffs on the same line plot
plt.figure(figsize=(12, 6))
for file, data in zip(name_list, relative_diffs.values()):
    plt.plot(data, label=file)
plt.legend()
plt.title('Relative Diffs over Time')
plt.xlabel('Prediction Step')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Plotting all relative errors as boxplots
plt.figure(figsize=(12, 6))
plt.boxplot([data for data in relative_errors.values()], labels=name_list, vert=True)
plt.title('Relative Errors')
plt.ylabel('Value')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```
